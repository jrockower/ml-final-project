```python
from arcgis.gis import GIS
from pathlib import Path
identifier = '7cb0877ba79b4c4f9a543ef161023854'
```

```python
anon_gis = GIS()
data_item = anon_gis.content.get(identifier)

data_path = Path('./data')

if not data_path.exists():
    data_path.mkdir()

data_item.download(save_path=data_path)
```
